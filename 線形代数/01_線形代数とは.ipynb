# 線形代数とは

## 線形代数$\neq$行列

**線形代数** を知っている人であれば「行列」を思い浮かべるかもしれません。それはおおよそ正しいです。
しかし、完全に一致するものではありません。

線形代数には **線形** という言葉が入っています。この「線形」は線形代数にとって非常にキーとなる言葉です。
数学での線形というのは次のように定義されています。

（とりあえず定義を紹介しますが「なんのこっちゃ」とでも思っておいてください。後の解説が重要です。）

#### 定義: 線形性

$A, B$を空でない集合とする。この集合$A$から$B$に移す**写像**を$f$とする。このとき、任意の$A$の要素$a,b$とスカラー$\alpha$に対して

$$
\begin{aligned}
f(a + b) &= f(a) + f(b)\\
f(\alpha a) &= \alpha f(a)\\
\end{aligned}
$$

を満たすとき、写像$f$は**線形である**という。


非常に数学じみた表現のため、わかりづらいかもしれません。また、説明していない言葉も出て来ています。軽く紹介します。

「集合」とは「要素の集まり」です。要素は数のようなものかもしれませんし、記号の集まりかもしれません。なんの集まりでも集合です。

「写像」とは、関数を抽象化したものです。線形代数の範疇では「関数」という認識で構いません。全章の説明で「関数は数と数をひもづけるもの」という説明をどこかでしていましたが、
写像も同じで、集合$A$の要素と集合$B$の要素を紐付けるものです。

「スカラー」とは単なる「数」です。変数ではなく、変数や関数を「○倍」したいと言うときに用いられる数値です。

さて、ここまで話がすんだところで、Juliaで線形である関数と線形でない関数を見ていきましょう。


```julia
# 線形である例
f = (x) -> 3 * x

# スカラー
α = 3.0

# 変数
a = 4.0
b = 10.0

println("f(a+b) = ", f(a+b))
println("f(a) + f(b) = ", f(a) + f(b))
println("f(a+b) == f(a) + f(b) ? ", f(a+b) == f(a) + f(b))
println("--------------------------")
println("f(α a) = ", f(α * a))
println("α f(a) = ", α * f(a))
println("f(α a) = α f(a) ? ", f(α * a) == α * f(a))
```

$f(x) = 3x$ という関数は$f(a + b) = f(a) + f(b)$と$f(\alpha a) = \alpha f(a)$がともに一致していました。ですので$f(x) = 3x$ は線形な関数のようです。

では次に、線形ではない関数を見てみましょう。

```julia
# 線形ではない例
g = (x) -> x^2

# スカラー
α = 3.0

# 変数
a = 4.0
b = 10.0

println("g(a+b) = ", g(a+b))
println("g(a) + g(b) = ", g(a) + g(b))
println("f(a+b) == g(a) + g(b) ? ", g(a+b) == g(a) + g(b))
println("--------------------------")
println("g(α a) = ", g(α * a))
println("α g(a) = ", α * g(a))
println("g(α a) = α g(a) ? ", g(α * a) == α * g(a))
```

$g(x) = x^2$ という関数の場合は、$g(a + b) = g(a) + g(b)$と$f(\alpha a) = \alpha g(a)$がともに一致していませんでした。ですので$g(x) = 3x$ は線形ではない関数のようです。

定義をみてもわかりづらいのですが、2つの条件の左辺は「変数を足したり掛けてから、関数を計算する」で右辺は「変数を関数を計算してから、足したり掛けたりする」となっています。
線形性というのは「計算する順番」に関わる性質のようです。

ではなぜ、「線形性」というのでしょうか。それは、グラフをみるとわかるかもしれません。

```julia
using Plots

f = x -> 3 * x
g = x -> x^2

p = plot(f, label="f(x) = 3x")
plot!(p, g, label="g(x) = x^2")
```

線形である$f(x) = 3x$は直線で、線形でない$g(x) = x^2$は曲線です。実は線形な関数というのは「直線である」という性質があります。

しかし「直線である関数はすべて線形である」といえるかというと、そうはいえません。次の例を見てみましょう。

```julia
# 線形ではない例
h = (x) -> 3x + 5

# スカラー
α = 3.0

# 変数
a = 4.0
b = 10.0

println("h(a+b) = ", h(a+b))
println("h(a) + h(b) = ", h(a) + h(b))
println("h(a+b) == h(a) + h(b) ? ", h(a+b) == h(a) + h(b))
println("--------------------------")
println("h(α a) = ", h(α * a))
println("α h(a) = ", α * h(a))
println("h(α a) = α h(a) ? ", h(α * a) == α * h(a))

```

$h(x) = 3x + 5$は線形ではないようです。また、さきほどの通りにグラフを見てみましょう。

```julia
using Plots

f = x -> 3 * x
h = x -> 3 * x + 5

x = -1:0.1:1

p = plot(x, f, label="f(x) = 3x")
plot!(p, x, h, label="h(x) = 3x + 5")
```

$f(x)$も$h(x)$も直線ですが、$f(x)$は線形で$g(x)$は線形ではありません。「何が違うのか」と問われても、おそらくこちらの性質はグラフからでは見えづらいです。

実は「関数が$x=0$のときに$f(x) = 0$となるか」という点にあります。

```julia
f = x -> 3x
h = x -> 3x + 5

　println("f(0) = ", f(0) )
　println("g(0) = ", h(0) )
```

グラフとしては、「関数が原点を通るか」がキーです。ちなみにこちらについては$g(0) = (0)^2 = 0$なので、$g(x)$は満たしていそうですね。

まとめると、「線形であること」とは「グラフが直線であり、原点を通る」という性質と同じようです。

と、「それがなんなんだ」と言う話ですね。「線形代数」の「代数」の方を解説し始めると大学の講義でも1年かかるので、とりあえずそちらは置いておきます。「線形性」というのを覚えておいてください。

で、これまでの説明では一切、行列がでてきていません。というか、写像（関数）の話しかしていません。なぜでしょうか？

実は**行列は関数と同様に、写像の一つである**と言ったら、どのように反応してもらえるでしょうか。驚いてもらえたら、とてもうれしいです。無反応が正しい姿かもしれません。

上の言葉のとおり、行列は写像の一つです。しかも行列は「線形性」を持つので**線形写像**と呼ばれるものの一つです。「行列は何か」ということをまだ紹介していませんが、
それは次の節に任せることにして行列の線形性をみていきます。

行列は大体、$A$のように大文字英語で表します。今の場合は$A$を行列としましょう。
関数のときのように、行列を「写像」と見なすのであれば、「集合から集合へ値を紐付けるもの」と考えるのはこれまでの説明からも自然な考えでしょう。
その集合の変数を$x,y$とします。あとで詳細を紹介しますが、一般的に$\bm{x}, \bm{y}$は**ベクトル**と呼ばれています。ベクトルは行列の「特殊な場合」です。
ベクトルは高校のときに習いますが、そのときは$\overrightarrow{x}, \overrightarrow{y}$のように表記していたかもしれません。
この書き方は幾何学ではよく使うことがありますが、多くの場合は太字で$\bm{x}, \bm{y}$、もしくは専門書になればあまり他の変数とは差をつけずに$\vec{x}, \vec{y}$と表記することも多いです。

いずれにしても「行列を写像と見なしたときの変数はベクトルである」ということを覚えておいてください。

また、これも関数のときと同様にスカラーとして$\alpha$を使います。これは関数のときと同じで、単なる数です。

このとき次のように言えます。


#### 定理: 行列の線形性

行列$A$に対して$A$と演算可能な任意のベクトルを$\bm{x}, \bm{y}$、スカラーを$\alpha$とする。このとき

$$
\begin{aligned}
A(x + y) &= Ax + Ay\\
A(\alpha x) &= \alpha A x
\end{aligned}
$$

が成り立つ。つまり行列は線形性を持つ。

表記は関数と異なりますが、関数のときと同様に「足したり掛けたりしてから、行列を計算」するのと「行列の計算をしてから、足したり掛けたりする」の両者は等しいことを意味しています。
ここでは行列とベクトルのたし算やかけ算、またはスカラー倍を明確にはしていません。しかし、この事実が成り立ちます。


これらが何を意味するかは徐々にわかっていきますが、ここで改めて、線形代数の話に戻りましょう。
線形代数では **線形空間** と **線形変換** という2つのことを最初に取り扱います。

線形空間は**ベクトル空間**とも呼ばれます。先ほどの行列の線形性で紹介したベクトルで構成される空間のことです。（空間、というものについてはどこかの節でコラムで紹介します。）
これについてはここで説明してもあまり意味を成さないですし、理解もしづらいので「ベクトルの集まり」くらいの意味で捉えてください。

線形変換とは「ベクトルを行列で書けることによって別の行列に移す」ものです。これは非常に視覚的に理解しやすいです。

今回は中学校3年生で習う「相似」な図形でしたが、行列を変えることで様々な図形にも変形できます。
それどころか、「」2次元から3次元」や「3次元から2次元」といった、次元間の変換もできます。

このような変形を線形変換と呼びます。要するに、行列をかける操作自体が「線形変換」というわけです。

線形代数では、このような線形空間や線形変換を学ぶ分野です。線形空間内にある要素であるベクトルを線形変換、ともいえるでしょうか。
この線形変換をするために「行列を掛ける」ため、行列について学ぶわけです。

しかし、現代ではこの限りではありません。そもそも、この行列が生まれたきっかけは紹介した線形変換というよりは、
連立方程式を解くための解法として考案された経緯があります。連立方程式とは、

$
\begin{aligned}
ax + by &= c\\
lx + my &= n\\
\end{aligned}
$

のように、複数の変数の式がいくつか並んでいる式のことです（詳細は別の節で紹介します）。連立方程式の解法や「そもそも解けるか？」という
ことを研究され、**行列式**というものが考案されます。行列式は、微分積分のさまざまな積分で、ヤコビアンの行列式として既に登場しています。

この行列式が長らく研究されるのですが、今から150年ほど前、19世紀のあたりで「行列式より行列の方をもっと研究すべきでは？」という人たちが声を上げ、
線形代数という分野ができました。それまでも行列自体は様々な応用の場面を見せていますが、行列の研究ができて行列そのものの性質がわかったことにより応用範囲が広がり、
現代では「数学の共通言語」の一つとして、数学（に限らず物理、工学、経済学など）の様々な分野で応用されるようになりました。

と、少々道がそれましたが、線形代数とは「線形空間や線形変換について学ぶものであり、それらには行列の扱いが重要である」として考えてもらえると、
単に「行列の計算が難しくてわかんねー」とはならずに済むと思い、解説してみました。
