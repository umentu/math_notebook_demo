# 三角関数

少々、難度高めです。微分には、これまでに紹介していない関数の微分が存在します。これらをすべて紹介すると専門書と変わらない内容になってしまうため、一例として三角関数の微分を紹介します。
その前に三角関数とはなにか、を少しだけ紹介します。三角関数は機械学習やフーリエ変換などでも出てくるため、知っておくとかなり有益です。

その前に**弧度法**を紹介します。小中学校では直角を$90^{\circ}$ といった表現で角度を表しますが、角度というのは計算する上で非常に不便です。
そのため「孤（円周の一部）の長さと角度を対比させて考える」のが弧度法です。

半径が1の円を考えます。すると円周の公式$2\pi r(rを円の半径とする)$より、円周の長さは$2\pi$です。これを基準に角度を表します。

円が一周したら角度は $360^{\circ}$です。このときの円周の長さは、$2 \pi$ でした。
半円のときの角度は $180^{\circ}$です。このときの孤の長さは、円周の半分なので$\pi$です。
四分円のときの角度は $90^{\circ}$です。このときの孤の長さは、半円の更に半分なので $\frac{\pi}{2}$です。
2周したら$4\pi$であり、0度は$0$です。

といったように表現することで、すべての角度を「長さ」で表現できます。

<img src="./images/07/01.png" alt="Drawing" style="width: 500px;"/>

弧度法により、三角関数を考えられるようになりました。


次に三角関数ですが、天下りで応用から説明します。

$xy$平面にある点を置きます。すると「座標」を考えられます。

<img src="./images/07/02.png" alt="Drawing" style="width: 500px;"/>

座標とは、「原点から各軸ごとの距離」としてみることができます。そのため$(x, y)$という形で表現できるわけです。
これを次の図のようにみることはできないでしょうか？


<img src="./images/07/03.png" alt="Drawing" style="width: 500px;"/>

原点と点を両端とする線分を引きます。この直線の長さを$r$、$x$軸と線分の間に挟まれた角度を$\theta$とします。
この2つの変数$(r, \theta)$で点を表現することができそうです。

すると、この$(x, y)$と$(r, \theta)$はどういった関係になるでしょうか。
これをうまく表現してくれるのが三角関数です。
実は$x, y$は美味い具合に$r,\theta$で次のように表現できます。

$$
\begin{aligned}
x &= r \cos \theta\\
y &= r \sin \theta\\
\end{aligned}
$$

なぜこのようにできるのか、というが何かということを見ることでわかります。

まず下の図をご覧ください。


<img src="./images/07/04.png" alt="Drawing" style="width: 500px;"/>

点$O$を中心とする円周上に点$A$をとります。点$O$と点$A$を両端とする線分を引きます。
次に（おそらく皆さんが苦手とする）補助線を3本引きます。1本目は点$A$から$x$軸に垂直に交差する線分（赤い線）を引きます。この線分と$x$軸と交差した点を$B$とします。
次に、OAを延長した直線を引きます。また、円と$x$軸が交差している点で$x$軸に垂直に交わるように直線を引きます（薄い青い線）。
青い直線同士が交差している点をC、$x$軸に垂直に交わっている点をDとします。

これで準備は完了です。

OAというのは円の半径ですので、ここでは半径を$r$としておきます。
更に、OAと$x$軸の間の角度を（弧度法で）$\theta$とします。このとき、sinとcosを次のように定義します。

$$
\begin{aligned}
\sin \theta &= \frac{\Delta OABの縦の長さ}{\Delta OABの斜辺の長さ}\\
            \\
            &= \frac{ABの長さ}{OAの長さ}\\
            \\
            &=  \frac{ABの長さ}{r}\\
\end{aligned}
$$

$$
\begin{aligned}
\cos \theta &= \frac{\Delta OABの横の長さ}{\Delta OABの斜辺の長さ}\\
                \\
            &= \frac{OBの長さ}{OAの長さ}\\
                \\
            &=  \frac{OBの長さ}{r}
\end{aligned}
$$


これがsinとcosの定義です。割りと多くの参考書ではここまでしか解説していませんが、もう一歩踏み込んでみましょう。
なぜ上のようにsinとcosを定義するのか、という点です。

sinは日本語で「正弦」、cosは日本語で「余弦」といいます。ここで疑問なのが「　正と余なのか」です。
$\Delta OAB$は$\angle ABO$が$90^{\circ}$なので直角三角形です。
三角形は内角の合計が$180^{\circ}$なので、$\angle ABO$を除いた二つの角の和は$90^{\circ}$です。
上の図のように三角形の直角部分を右下に持ってきたときに、「底辺から反時計周りに回転させた角」、つまり$\angle AOB$の角を**正角**といい、
最後に残っている角、すなわち$\angle OAB$を**余角**といいます。
また、正角の反対側にある辺ABを「正角に対応する辺」とし、余角に対応する辺OBを「余角に対応する辺」とします。

!()[./images/]

sinは$\frac{AB}{OA}$でした。これは言葉で置き換えると「「正角に対応する辺」と斜辺の比率」です。つまりsinは正角に関する辺の比率を考えているので「正弦」です。

cosは$\frac{OB}{OA}$でした。これは言葉で置き換えると「「余角に対応する辺」と斜辺の比率」です。つまりcosは余角に関する辺の比率を考えているので「余弦」です。






では、なぜこれが先ほどの座標の話に行き着くのでしょうか。
上の定義を少し変形してみましょう。

$$
\begin{aligned}
\sin \theta &= \frac{ABの長さ}{r}\\
                \\
ABの長さ　     &= r \times \sin \theta
\end{aligned}
$$

$$
\begin{aligned}
\cos \theta &= \frac{OBの長さ}{r}\\
                \\
OBの長さ　     &= r \times \cos \theta
\end{aligned}
$$m

線分ABの長さと線分OBの長さを表す式となりました。右辺はどこかで見覚えがありますね。
先ほどの

$$
\begin{aligned}
x &= r \cos \theta\\
y &= r \sin \theta\\
\end{aligned}
$$

でした。
実はABの長さは$\Delta OAB$の縦の長さでした。更に点$O$は原点でしたので、ABの長さは「原点からの$y$軸上の距離」と見なせます。
OBも同様に「原点からの$x$軸上の距離」と見なすことができます。
先ほどのグラフで表すと、次のようになります。

<img src="./images/07/06.png" alt="Drawing" style="width: 500px;"/>

$xy$平面上の点は円の大きさを変えれば「三角形の斜辺の長さ」　はいくらでも変えられます。
$\theta$については、最大でも円を一周するときの角度と考えられるため、$0 \lte \theta　\lte \pi$の範囲で動きます。
つまり、座標上のあらゆる点は$r, \theta$の二つの変数で表現できる、ということになります。

「$(x, y)$ではなく$(r, \theta)$がなぜ必要なのか？」と思うかもしれません。（「なぜ必要なのか」というのはナンセンスな質問ですが、人は思いがちです。）
必要性についてはたくさんありますが、ここでは「そう表現することでできるようになる微分や積分の計算があるから」としておきます。


さて、先ほどの補助線を引いたときに、線分CDが出て来ました。これはまだ登場していません。
この補助線はtanを紹介するためのものです。

先にtanの定義を書きます。

$$
\begin{aligned}
\tan {\theta} &= \frac{\sin \theta}{\cos \theta}\\
              \\
              &= \frac{\frac{AB}{OA}}{\frac{OB}{OA}}\\
              \\              
              &= \frac{AB}{OB}
\end{aligned}
$$

今度は「sinとcos」の比率のようです。sin,cosともに分母に斜辺の長さがあるので、比率を考えると相殺されて、結局、「「正角に対応する辺」と「「余角に対応する辺」の比率を考えています。
定義としてはこのようになっていますが、それがなぜ線分CDが関係するのでしょうか。

理由は、次の図でわかるようになるかもしれません。

<img src="./images/07/08.png" alt="Drawing" style="width: 500px;"/>

$\Delta OAB$と$\Delta OCD$の底辺の長さと高さの長さの比率を考えています。ここでは分かりやすく、ODの長さを1としています。

$$
\begin{aligned}
AB:OB  &= CD:OD\\
AB \times OD  &= OB \times CD\\
AB \times 1  &= OB \times CD\\
AB   &= OB \times CD\\
CD    &= \frac{AB}{OB}
\end{aligned}
$$

tanの定義を見返してもらえればご覧の通り、$\tan \theta$と$CD$の右辺が一致しました。これが接線を引いた意味です。
ちなみに、tanは日本語で**正接**といいます。

さて、45度（弧度法で$\frac(1){4}\pi$）ごとのsin,cos,tanを見てみましょう。


```julia
for theta=0:(1/8)pi:4pi
    println("角度が",floor(theta, digits=3),"のとき、sinは",floor(sin(theta), digits=3),"、cosは",floor(cos(theta), digits=3),"、tanは",floor(tan(theta), digits=3))
    if theta == 2pi
        println("-----------1週目終わり---------")
    end
end
```

sinだけに注目すると、sinは最大で1で最小で-1です。
cosも同様に、sinは最大で1で最小で-1です。
tanはさまざまな数が出て来ています。「1.633123935319537e16」と言った数は膨大に大きい数を表していますが、これは丸めているだけで実質は$\infty$です。

グラフでも見てみましょう。

```julia
using Plots

p = plot(0:0.1:2pi, sin, label="sin" )
plot!(p, 0:0.1:2pi, cos, label="cos")
```

よくみる周波数の図ですね。これは範囲を絞っていますが、ずっとこのような波型で周期します。
見方を変えると上で紹介した通り、sinは円周上の$y$座標、cosは円周上の$x$座標をぐるぐる回っているだけなので、このような周期性であるのは当然と考えられます。

続いてtanを見てみましょう。

```julia
using Plots
plot(-10pi:0.1:10pi, tan, ylim=[-10, 10])
```

心電図のようなグラフができました。この上にも下にも飛び跳ねているところは、それぞれ$\infty, -\infty$に発散している部分です。
これは$(\frac{1}{2}+n)\pi \quad (n=0,\pm 1, \pm 2, \pm 3, \dots)$ごとに現れます。
これはtanの定義である$cos$が$0$に近づいたときに起こります。上の数値で各三角関数の値をみると$cos=0.0$のときに起こっていることがみてとれますね。

三角関数についてはたくさんの性質があるのですが、それらは利用しながら紹介するので、弧度法と三角関数についてはこれで終わります。
