# 微分と積分の諸性質

## 速度で見る微分と積分

くどいですが、今までの説明では微分は「ある瞬間の変化率」で積分は「ある区間の図形の面積」でした。
そして「微分積分学の基本定理」によって、微分と積分は相反する演算であることが明らかになりました。

これを応用すると、さまざまな性質がみえてきます。ここでは一度、「速度」に置き換えて説明します。

とある乗り物の進んだ距離を計測します。すると$t$秒後に$f(t) = t^3$の地点にいることがわかりました。
するとグラフは次のようになります。


````julia
using Plots
function distance(t)
    return t^3
end

plot(distance, 0, 10)
````

横軸が時間で縦軸がその地点でいるスタート地点からの距離です。$f(t)$を微分してみます。

<img src="./images/04/01.png" alt="Drawing" style="width: 500px;"/>

微分してみると、　$f'(t) = 3t^2$になりました。この$t$に具体的な時間$t_1$を入れるとその瞬間の変化率、すなわち$f(t_1)$の地点の接線の傾きが求められます。
では、この「接線の傾き」は何を意味するでしょうか。今回の場合は「速度」になります。
上のグラフをみてみると、時間を経過するごとに急激にスタート地点からの距離が伸びているのがわかります。これは時間が経過するごとに「速度」が増えているためです。
具体的に速度を計算してみましょう。

````julia
function speed(t)
    return 3t^2
end

for t=0:10
    println("時間t=",t,"の速度は",speed(t))
end
````

速度が徐々に上昇していることがわかります。次に、$f(t)$と各時間における接線の関係をみてみましょう。

````julia
function speed(t)
    return 3t^2
end

tan01 = x -> speed(1) * (x-1) + distance(1)
tan02 = x -> speed(2) * (x-2) + distance(2)
tan03 = x -> speed(3) * (x-3) + distance(3)
tan04 = x -> speed(4) * (x-4) + distance(4)
tan05 = x -> speed(5) * (x-5) + distance(5)
tan06 = x -> speed(6) * (x-6) + distance(6)
tan07 = x -> speed(7) * (x-7) + distance(7)
tan08 = x -> speed(8) * (x-8) + distance(8)
tan09 = x -> speed(9) * (x-9) + distance(9)
tan10 = x -> speed(10) * (x-10) + distance(10)

plt = plot(distance, label="distance", 0, 10,  legend = :bottomright, ylim=(0, 1000), lw=4)
plot!(plt, tan01, label="tan01")
plot!(plt, tan02, label="tan02")
plot!(plt, tan03, label="tan03")
plot!(plt, tan04, label="tan04")
plot!(plt, tan05, label="tan05")
plot!(plt, tan06, label="tan06")
plot!(plt, tan07, label="tan07")
plot!(plt, tan08, label="tan08")
plot!(plt, tan09, label="tan09")
plot!(plt, tan10, label="tan10")
````

1秒ごとの接線をみていくと、徐々に傾きが強くなっていることがみてとれます。ちなみに接線の関数の求め方は、$t_1$秒時の座標を$(t_1, f(t_1))$なので、

$$
\begin{aligned}
y-f(t_1) &= f(t_1) * (t-t_1) \\
y &= f(t_1)(t-t_1) + f(t_1)    
\end{aligned}
$$

として求められます。

さて、速度の関数は$f'(t) = 3t^2$ でした。これをグラフにすると次のようになります。

````julia
f_speed = t -> 3t^2

plot(f_speed, 0, 10)
````

$f(t) = t^3$ 似たようなグラフですが異なります。これを再度微分してみましょう。

<img src="./images/04/02.png" alt="Drawing" style="width: 500px;"/>

すると、$f''(t) = 6t$となりました。$f''(t)$は時間$t$の瞬間の速度の変化率なので「加速度」です。つまり、 速度を微分すると「加速度」になります。
時間単位の加速度を見てみましょう。

````julia
function acceleration(t)
    return 6t
end

    

for t=0:10
    println("時間t=",t,"の加速度は",acceleration(t))
end
````

$t$が1増えるごとに加速度は6ずつ単調に上がっています。これはグラフでも確認できます。

````julia
plot(acceleration, 0, 10)
````

速度と加速度の関係を見てみるために、再度接線を描いてみましょう。

````julia
tan11 = x -> acceleration(1) * (x-1) + speed(1)
tan12 = x -> acceleration(2) * (x-2) + speed(2)
tan13 = x -> acceleration(3) * (x-3) + speed(3)
tan14 = x -> acceleration(4) * (x-4) + speed(4)
tan15 = x -> acceleration(5) * (x-5) + speed(5)
tan16 = x -> acceleration(6) * (x-6) + speed(6)
tan17 = x -> acceleration(7) * (x-7) + speed(7)
tan18 = x -> acceleration(8) * (x-8) + speed(8)
tan19 = x -> acceleration(9) * (x-9) + speed(9)
tan20 = x -> acceleration(10) * (x-10) + speed(10)

plt = plot(speed, label="speed", 0, 10,  legend = :bottomright, ylim=(0, 300), lw=4)
plot!(plt, tan11, label="tan11")
plot!(plt, tan12, label="tan12")
plot!(plt, tan13, label="tan13")
plot!(plt, tan14, label="tan14")
plot!(plt, tan15, label="tan15")
plot!(plt, tan16, label="tan16")
plot!(plt, tan17, label="tan17")
plot!(plt, tan18, label="tan18")
plot!(plt, tan19, label="tan19")
plot!(plt, tan20, label="tan20")
````

黙視だと見えづらいかもしれませんが、距離と速度の比較に比べると接線が緩やかなことが見て取れます。

可視化はしませんが、加速度を微分すると$f^{(3)}(t) = 6$となります。これは加速度の瞬間の変化率ですが、$t$に何を入れても$6$です。つまり加速度の変化率は常に一定です。
これまでのことを逆にみなすと、

```
加速度の変化率が6で一定である。 
-> 速度の変化率である加速度が常に6ずつ一定に速度は6tずつ一定に上昇する。
-> 距離の変化率である速度が時間ごとに3t^2ずつ上昇していくから距離はt^3ごとに急激に上昇する。
```

ように言い換えることもできます。微分することで、関数自体がどのように変化していくかを捉えられます。

また、積分は微分の逆の演算でしたので、加速度を積分すると速度の関数、速度を積分すると距離の関数を求められます。



##　注意点

最初に、積分には**定積分**と**不定積分**というものがあります。機械学習で出てくる積分は定積分がほとんどのため、わからなければ飛ばしてください。

定積分は「積分する範囲が決まっている」積分です。


$$ \int_{a}^{b} f(x) dx　　$$

それに対して不定積分は「積分する範囲が決まっていない」積分です。

$$ \int f(x) dx or F(x) = \int_{a}^{x} f(x) dx $$

機械学習での積分は後で紹介するように、確率を面積で考えて計算するため、「ある条件からある条件に入る面積」を求めることが多いです。
そのため、以降で説明する積分定数を考える必要がありません。

しかし、関数自体の原始関数を知りたい場合には不定積分をします。
その際に注意をする必要があります。


今までの計算では出てこなかったため、ここで改めて**積分定数**というものを紹介します。
$f(x) = 3$ のように、$x$に何を入れても変化しない数を **定数関数** といいます。
定数関数は$x$について微分しても、必ず$0$になります。

$$f'(x) = 0$$

また、$f(x) = x^2 + 3x - 4$ のように項($x^2$, 3x, -4)が複数ある場合には、項ごとに微分します。

$$f'(x) = 2x + 3$$

項ごとに微分をするため、$-4$は定数関数と同じものとみなして$0$になります。

……ここまで大丈夫でしょうか。ここで $g(x) = x^2 + 3x + 10$という関数を見てみます。これは明らかに$f(x)$ と異なる関数ですが、

$$g'(x) = 2x + 3$$

と微分すると$f'(x)$と等しくなります。しかし先ほど確認した通り、$f(x)$と$g(x)$は異なる関数です。

さて、微分の反対の演算は積分でした。なので$f'(x)$を積分したら$f(x)$に、$g'（x）$を積分したら$g(x)$になってほしいところですが$f'(x)$と$g'(x)$は同じ関数なので、積分した結果も等しくなりそうです。

実は、不定積分の場合は必ず「見えざる定数」が出現します。$f'(x)$を不定積分した場合には、

$$ \int f'(x) dx = f(x) + C$$

$C$は＄x＄とは関係のない、単なる数値です。そのため微分すれば消えます。

$$ (\int f'(x) dx)' = f'(x) $$

実際の計算ではあまり出てこないですが、証明などで不定積分を使う場合には注意が必要です。



## 積分と確率

確率を真剣に説明しようとすると統計学の範囲となるため、雑に説明します。

機械学習で確率を計算するときには「面積」で考えます。
確率は0から1です。ですので、積分した結果が$0.5$　であれば確率は50%のように考えます。

正規分布、もしくはガウス分布と呼ばれる確率モデルがあります。これは次のようなグラフです。

````julia
# 平均
mu = 0.0
# 標準偏差
sigma = 1.0


function normal_distribution(x) 
    # exp() はネイピア数 e=2.718... の指数関数です。
    return 1 / sqrt(2 * pi * sigma^2) * exp(-(x-mu)^2/(2 * sigma^2))
end

plot(normal_distribution, label="f(x)")
````

正規分布がどのように使われるかはここではお話しませんが、正規分布はすべての範囲の面積を足すと$1$になる、という特性があります。
すなわち、

$$ \int_{- \infty}^{\infty} f(x) dx = 1 $$

機械学習では確率モデルがたくさん出て来ますが、すべての確率モデルはすべての範囲の面積を足すと$1$になるように作られています。
そのため、「面積を求めること」と「確率を求めること」の対応付けができるわけです。

試しに$x=-2$から$x=2$までの面積の近似を区分求積法で求めてみましょう。

````julia
N = 10000
x = -2.0

result = 0.0

for n=0:N
    result += (4/N) * normal_distribution(n*4/N + x)
end

println(result)
````
